## Formatted String (f-string)

```python
"""
In case we want to change a string,
we put it into a formatted string
"""
f_temp = 21
f'The temperature is {f_temp} degrees Celcius'
```

With this we can make changes on it.

```python
f'The temperature is {(f_temp * 1.8) + 32} degrees Fahrenheit'
```

### Rounding with f-strings


```python
f'The temperature is {(f_temp * 1.8) + 32:.2f} degrees Fahrenheit'
```

```python
f'The temperature is {(f_temp * 1.8) + 32:.2e} degrees Fahrenheit'
```

```python
f'The temperature is {(f_temp * 1.8) + 32:.2%} degrees Fahrenheit'
```

### Big numbers with f-strings

```python
f_temp_sun = 27000000
```

```python
f'The temperature is {(f_temp_sun * 1.8) + 32:.2f} degrees Fahrenheit'
```

```python
f'The temperature is {(f_temp_sun * 1.8) + 32:,.2f} degrees Fahrenheit'
```

### More formatting with f-strings

```python
f_temp_earth = 21

print(f'The temperature is {(f_temp_sun * 1.8) + 32:,.2f} degrees Fahrenheit')
print(f'The temperature is {(f_temp_earth * 1.8) + 32:,.2f} degrees Fahrenheit')
```

:[width],.[decimals][type]

```python
print(f'The temperature is {(f_temp_sun * 1.8) + 32:15,.2f} degrees Fahrenheit')
print(f'The temperature is {(f_temp_earth * 1.8) + 32:15,.2f} degrees Fahrenheit')
```

We set [width] as selecting $15$, which helps the readability, now each value: $48,600,032.00$ and $69.80$ has the same width (space): $15$.

:[alignment][width],.[decimals][type]

```python
print(f'The temperature is {(f_temp_sun * 1.8) + 32:<15,.2f} degrees Fahrenheit')
print(f'The temperature is {(f_temp_earth * 1.8) + 32:<15,.2f} degrees Fahrenheit')
```

Aligned them to left.

```python
print(f'The temperature is {(f_temp_sun * 1.8) + 32:^15,.2f} degrees Fahrenheit')
print(f'The temperature is {(f_temp_earth * 1.8) + 32:^15,.2f} degrees Fahrenheit')
```

Aligned them to centre.

```python
print(f'The temperature is {(f_temp_sun * 1.8) + 32:>15,.2f} degrees Fahrenheit')
print(f'The temperature is {(f_temp_earth * 1.8) + 32:>15,.2f} degrees Fahrenheit')
```

Aligned them to right again.

:[fill][alignment][width],.[decimals][type]

```python
print(f'The temperature is {(f_temp_sun * 1.8) + 32:*^15,.2f} degrees Fahrenheit')
print(f'The temperature is {(f_temp_earth * 1.8) + 32:*^15,.2f} degrees Fahrenheit')
```

Filled empty spaces with `*`.
