## Elif & Logical Expressions

```python
temperature = 55
is_cold = temperature < 50

if is_cold:
  print("Wear a jacket!")
else:
  print("No jacket")

print("End of program")
```

Let's add a third condition via `elif`:

```python
temperature = 45
is_cold = temperature < 50

if is_cold:
  print("Wear a jacket!")

elif temperature < 65:
  print("Some nonsense")

else:
  print("No jacket")

print("End of program")
```

A more clear example:

```python
temperature = 45

if temperature < 32:
  print("Wear a jacket!")

elif temperature < 50:
  print("Getting hotter")

elif temperature < 65:
  print("Even more")

elif temperature < 80:
  print("Crazy hot!")

else:
  print("No jacket")

print("End of program")
```

Let's imagine we are increasing number of conditions by the settings:

```python
temperature = 45    # we set temperature to an integer
is_raining = True   # we set is_raining to a boolean

"""
Two conditions:
temperature < 50
is_raining == True
Merge two conditions via "and" a logical operator
"""
temperature < 50 and is_raining == True   # temperature < 50: True, is_raining = True
```

Here, we end up with `True` only when both conditions are met! which we set them beforehand  and `True`.

```python
temperature = 45
is_raining = False

temperature < 50 and is_raining == True # # temperature < 50: True, is_raining = False
```

### Logical experssion: `and`

|Cond1|Cond2|Expression|
|---|---|---|
|`True`|`True`|`True`|
|`True`|`False`|`False`|
|`False`|`True`|`False`|
|`False`|`False`|`False`|

```python
# True and True
temperature = 45
is_raining = True

if temperature < 50 and is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

```python
# True and False
temperature = 45
is_raining = False

if temperature < 50 and is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

```python
# False and True
temperature = 55
is_raining = True

if temperature < 50 and is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

```python
# False and False
temperature = 55
is_raining = False

if temperature < 50 and is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

### Logical experssion: `or`

|Cond1|Cond2|Expression|
|---|---|---|
|`True`|`True`|`True`|
|`True`|`False`|`True`|
|`False`|`True`|`True`|
|`False`|`False`|`False`|

```python
# True or True
temperature = 45
is_raining = True

if temperature < 50 or is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

```python
# True or False
temperature = 45
is_raining = False

if temperature < 50 or is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

```python
# False or True
temperature = 55
is_raining = True

if temperature < 50 or is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

```python
# False or False
temperature = 55
is_raining = False

if temperature < 50 or is_raining:
  print("Wear a warm rain jacket!")

print("End of program")
```

### Logical experssion: `not`

|Cond|Expression|
|---|---|

```python
is_raining = False
not is_raining
```

```python
is_raining = True
temperature = 75

temperature > 70 and not is_raining
```
