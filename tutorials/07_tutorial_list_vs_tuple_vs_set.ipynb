## Lists vs Tuples vs Sets
`[]Lists`, `()Tuple`, `{}Set`

```python
# List
['a', 'b', 'c', 'a']
```

**Keep** duplicates, **ordered**

```python
# Tuple
('a', 'b', 'c', 'a')
```

**Keep** duplicates, **ordered**

```python
# Set
{'a', 'b', 'c', 'a'}
```

Immediately we see that `set` **drops** duplicates, **unordered**

```python
nums = [1, 3, 3, 1, 2, 4, 5, 2, 4]
list(set(nums))
```

```python
my_list = ['a', 'b', 'c', 'a']
my_list[2]
```

```python
my_tuple = ('a', 'b', 'c', 'a')
my_tuple[2]
```

```python
my_set = {'a', 'b', 'c', 'a'}
my_set[2]
```

Because `set` is **unordered**; no index

### List vs Tuple
List is **mutable**, Tuple is **immutable**


```python
# List
my_list.append('d')
my_list
```

```python
# Tuple
my_tuple.append('d')
my_tuple
```

`Tuple` can be used as a tool for **accidental** changes. For example, it can be used for data that should not be changed like graduation date.

`Set` is **mutable**

```python
# Set
# Change my_set from {'a', 'b', 'c'} to {}'a', 'b', 'c', 'd'}
my_set = {'a', 'b', 'c', 'd'}

# Remove an item 'a'
my_set.remove('a')
my_set
```

| |List|Tuple|Set|
|---|---|---|---|
|Syntax|[]|()|{}|
|Dublicates?|Yes|Yes|No|
|Ordered?|Yes|Yes|No|
|Mutable?|Yes|No|Yes|
