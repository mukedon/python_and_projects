## Dictionaries
dictionary = {key:value}

```python
contacts = {
    'Richard': 'rich@gmail.com',
    'Sibel': 'cybl@hotmail.com',
    'Ron': 'ron@hogwartsmail.com'
}

print(contacts['Ron'])
print(contacts['Richard'])
print(contacts['Sibel'])
```

```python
contacts
```

Dictionaries

    items defined by key
    ideal for data with meaningful labels

Lists

    items defined by position

An example of Dictionary
|Key|Value|
|---|---|
|Usernames|Passwords|
|Products|Prices|
|Students|Grades|

**Key & value rules**
1. Values can be any type
2. Multiple keys can have the same value
3. Keys must be unique
4. Keys don't need to be strings

```python
# Adding an item to a dictionary
contacts['Alex'] = 'alex@gmail.com'
contacts
```

```python
# Removing an item from the dictionary
contacts.pop('Ron')
contacts
```

Dictionary methods:

Inquiry on a key; if it is in the dictionary or not; use `.keys()` method

```python
contacts.keys()
```

```python
for key in contacts.keys():
  if key == 'Alex':
    print('Alex is in contacts')
```

Inquiry on a value; if it is in the dictionary or not; use `.values()` method

```python
contacts.values()
```

```python
for value in contacts.values():
  if value == 'alex@gmail.com':
    print('Email found')
```

Which value matches which key? use `.items()`

```python
contacts.items()
```

```python
for key, value in contacts.items():
  if value == 'alex@gmail.com':
    print(value, '->', key)
```
