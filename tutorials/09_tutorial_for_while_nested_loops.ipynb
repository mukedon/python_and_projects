## For Loops
Iteration without condition

```python
discount = 0.10
prices = [12.99, 15.99, 7.99, 27.99]
```

```python
print(prices[0]*(1-discount))
print(prices[1]*(1-discount))
print(prices[2]*(1-discount))
print(prices[3]*(1-discount))
```

```python
for price in prices:
  print(f"Price {price} is discounted to:", price * (1 - discount))
print("Prices discounted!")
```

```python
ticket_numbers = [0, 1, 2, 3, 4, 5]

for ticket_number in ticket_numbers:
  print("Summer's super ruffle")
  print("Ticket number:", ticket_number)
```

```python
for ticket_number in range(100, 106):
  print("Summer's super ruffle")
  print("Ticket number:", ticket_number)
```

## While Loops
Conditional iteration

```python
count = 1

while count <= 5:
  print("count =", count)
  count += 1

print('End of the program')
```

```python
count = 1

while count <= 5:
  print("count =", count)
  count += 2

print('End of the program')
```

### While vs For loops

```python
for count in range(1, 6):
  print('count =', count)

print('End of program')
```

```python
user_input = ''

while user_input != 'secret123':
  user_input = input('Enter your password: ')

print('Access granted!')
```

### While `True`syntax

```python
while True:
  user_input = input('Enter your password: ')

  if user_input == 'secret123':
    break

print('Access granted!')
```

**Summary**
1. **Syntax & Execution**

    `while` **condition** == `True`:

        code to repeat

2. `while` loops vs `for` loops

    `for`: # of iterations is known

    `while`: # of iterations is unkown

3. `while` `True:

       Syntax

    `if` should_end:

      `break`

### Nested Loops

```python
numbers = [1, 2, 3]
letters = ['a', 'b', 'c']
```

```python
for number in numbers:
  print(f"{number}{letters[0]}", end=" ")
  print(f"{number}{letters[1]}", end=" ")
  print(f"{number}{letters[2]}", end=" ")
  print()
```

Use Nested loop instead

```python
for number in numbers:
  for letter in letters:
    print(f"{number}{letters[0]}", end=" ")
  print()
```
