### Math Operators:
* `+` : sum
* `-` : subtract
* `*` : multiply
* `/` : divide
* `**` : exponent
* `//` : floor division
* `%` : modulus

```python
print(2 + 3)
print(8 - 3)
print(5 * 7)
print(9 / 4)
print(2 ** 3)
print(8 // 3)
print(8 % 3)
```
