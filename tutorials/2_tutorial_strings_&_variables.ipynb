### Strings
They account for texts in the code and they can be used for myriad of cases; input, outputs, ...

* They can be entered via `''` or `""`.

```python
"""
'' or "" give the same output.
Inside '' or "" letters, punctuation (.?!...)
symbols (* & % @ ...) numbers (123...) can be entered
(for numbers they lose they integer, float, ... abilities)
"""
print('cat')
print("cat")
```

```python
print(1 + 2)
```

```python
print('1' + '2')
```

It is obvious that, in the first print python sums up two integers meanwhile in the second print it takes them as strings and sums up these strings as putting them side by side in output.

* print(1 + 2): integers → `addition`
* print('1' + '2'): strings → `concatenation`

```python
print(1 + '2')
```

As expected, summation of integer and string is not supported, because they belong to different class types and hence it give TypeError (here for mixing different types for the given function `print`).

```python
type(1)
```

```python
type('2 ')
```

```python
print(00123)
```

Here python sees $00123$ as integer and as we know $00123$ is not an integer (at least in the sense that is defined in python) and hence outputs a SyntaxError.

```python
print('00123')
```

Here we again used $00123$ and no error is face, because this time it defined as string and for string a number $00123$ is not a problem for python.

```python
print('Bob's cat')
```

```python
print("Bob's cat")
```

By the two examples above using `''` for one item is more suitable, meanwhile for more than one item `""` is more convenient.
