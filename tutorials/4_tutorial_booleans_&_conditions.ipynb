## Booleans and Conditionals
Code to make decisions
1. **Ask/answer** yes or no questions
2. **Run different code** based on question's answer.

* **Question**: Is it cold outside?
* **Answer**: Boolean → **Yes (True)** or **No (False)**

```python
temperature = 45
temperature < 50
```

```python
temperature = 45
temperature > 50
```

Here, we set temperature as $50$ and then we check if it is less than $50$ via `<`. Correctly, it gives boolean: `True` because temperature is less than $50$ as it is $45 < 50$. Similarly, we check if it is more than $50$ via `>` and correctly it gives boolean: `False` because temperature is not more than $50$ as it is $45 \ngtr 50$. We used $>, <$ comparison operators here.

```python
temperature = 55
is_cold = temperature < 50 # which will give false
is_cold
```

### Operators:
* `<`: less than
* `>`: greater than
* `<=`: less or equal
* `>=`: greater or equal
* `==`: equal
* `!=`: not equal

```python
temperature = 45
is_cold = temperature < 50

if is_cold:
  print("Wear a jacket!")

print("End of program")
```

```python
temperature = 55
is_cold = temperature < 50

if is_cold:
  print("Wear a jacket!")

print("End of program")
```

In first `if` statement, because temperature is 45, print function in the `if` is executed.

In second `if` statement, because temperature is 55, print function in the `if` is _not_ executed.

```python
temperature = 55
is_cold = temperature < 50

if is_cold:
  print("Wear a jacket!")
else:
  print("No jacket")

print("End of program")
```

Here, we checked both `if` statement by its both booleans: `True` would print first statement, `False` would print second.
