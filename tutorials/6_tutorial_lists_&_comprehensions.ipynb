## Lists
A way of storing data in python, that can be changed later.

Example: a list of friends stored in list called: 'friends'.

```python
friends = ["Richard", "Sibel", "Nathan", "William"]
print(friends)
```

Items in the list by their indexes; Richard: `0`, Sibel: `1`, Nathan: `2`, William: `3`.

```python
# Calling an item in the list by its index
print(friends[0])
```

```python
print("Hello", friends[1])
```

```python
# Calling an item in the list by its index and reassign it
friends[2] = "Nate"
print(friends)
```

```python
print("Hello", friends[2])
```

```python
# Calling specific items in the list by their indexes
# While this calling python includes first parameter, excludes second
# my_list[start_index (include):end_index (exclude)]

# Between indexes 1 and 2 = only index 1
print(friends[1:2])

# Between indexes 1 and 3 = 1 and 2 (with 1 step by default)
print(friends[1:3])

print(friends[1:])

# Negative indexing
print(friends[-3:-1])
print(friends[-3:])
```

```python
# Add an item to the list; use method: .append()
friends.append("Jin")
print(friends)
```

```python
# Remove an item from the list use method: .remove()
friends.remove("Nate")
print(friends)
```

```python
# Rearrange list alphabetically; use method: .sort()
friends.sort()
print(friends)
```

```python
# When you don't remember the useable methods,
# After the list name (or any other storage) just put `.`
# To check the all allowed methods
friends.
```

```python
# Creating two lists
home_friends = ["Sibel", "Jin", "Dana"]
work_friends = ["Richard", "Nate", "William"]

# Combining lists
all_friends = home_friends + work_friends

# Sorting the list
all_friends.sort()

# Printing the list
print(all_friends)
```

### List Comprehensions
Example: [n**2 for n in nums]

```python
# Create a list of number
nums = [1, 2, 3, 4, 5]

# Create an empty list to fill
nums_squared = []

# Run a for loop for items in nums
for n in nums:
  # Define variable: square as equal
  # to square of items (integers in this list)
  square = n**2

  # add each squared items into nums_square
  nums_squared.append(square)

# Print nums_square
print(nums_squared)
```

```python
# Same process with list comprehension
# [operation | for var in | orig_list]
nums_squared = [n**2 for n in nums]
print(nums_squared)
```

```python
tv_shows = ["Seinfeld", "the Office",
            "PARKS AND RECREATION", "modern Family", "30 rock"]

# Capitalize with .title()
tv_shows_cap = []
for show in tv_shows:
  show_cap = show.title()
  tv_shows_cap.append(show_cap)

print(tv_shows_cap)
```

```python
# Now with comprehension
tv_shows_cap = [show.title() for show in tv_shows]
print(tv_shows_cap)
```

### Working with conditions

```python
# Use condition: if len(show) >= 10:
tv_shows_cap = []
for show in tv_shows:
  if len(show) >= 10:
    show_cap = show.title()
    tv_shows_cap.append(show_cap)

print(tv_shows_cap)
```

```python
# Now with comprehension again
tv_shows_cap = [show.title() for show in tv_shows if len(show) >= 10]
print(tv_shows_cap)
```

**List comprehensions**
* skip `.append()` call
* executed as a single, **optimized** step

**Use when**
* Logic is **simple**
* Need to **transform** or **filter**

```python
nums = range(1, 100000001)
```

```python
result = []
for n in nums:
  if n % 2 == 0:
    square = n**2
    if square > 10:
      result.append(square)
result
```

```python
result = [n**2 for n in nums if n % 2 == 0 and n**2 > 10]
result
```
