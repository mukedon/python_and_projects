Using a `for` loop:

```python
numbers = [1, 2, 3, 4, 5]
doubled_numbers_for = []
for num in numbers:
  doubled_numbers_for.append(num*2)
print(f"Doubled numbers (for loop): {doubled_numbers_for}")
```
